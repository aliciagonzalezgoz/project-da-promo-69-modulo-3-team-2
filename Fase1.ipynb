## Pregunta 1 - ¿Qué porcentaje de empleados abandona la empresa? 

`Attrition` es la columna más importante del dataset. Indica si el empleado abandonó la empresa (`Yes`) o sigue en ella (`No`). 

Para el proyecto de cursos de idiomas, **solo nos interesan los empleados con `Attrition = No`**. Ofrecer un curso caro a alguien que ya se fue o que tiene alta probabilidad de irse sería un desperdicio de recursos.

```python
print("=== DISTRIBUCIÓN DE ATTRITION ===\n")
print(df['Attrition'].value_counts())
print()
print("Porcentaje:")
print((df['Attrition'].value_counts(normalize=True) * 100).round(1).astype(str) + '%')
print()

# Universo de candidatos potenciales
candidatos = df[df['Attrition'] == 'No']
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
print(f"Total empleados en el dataset:       {len(df):>5}")
print(f"Empleados que se fueron (excluidos): {df['Attrition'].eq('Yes').sum():>5}")
print(f"Candidatos potenciales al curso:     {len(candidatos):>5}")
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
```

### PREGUNTA3. ¿Qué puestos tienen más riesgo de rotación?  JobRole, Attrition  Identificar roles con mayor salida. 
 
- Ver todos los puestos con más rotación, para saber cuales tiene más tasa de abandono:

```python
#PREGUNTA3. QUE PUESTOS TIENE MAS RIESGO DE ROTACION?
 
#Ver todos los puestos con más rotación, para saber cuales tiene mas tasa d abandono:
 
df[df["Attrition"] == "Yes"]["JobRole"].value_counts().reset_index()  #calcula por volumen y no por tasa, puede llevar a conclusiones erroneas
```

```python
#para calcular la tasa de rotación por puesto, necesitamos el número de empleados que se fueron por cada puesto (absolutos) y el número total de empleados por cada puesto (total).
absolutos = df[df["Attrition"] == "Yes"]["JobRole"].value_counts().reset_index()
absolutos.columns = ["JobRole", "Bajas"]

# Total de empleados por puesto
total = df["JobRole"].value_counts().reset_index()
total.columns = ["JobRole", "Total"]

# Unir y calcular tasa de rotación
rotacion = absolutos.merge(total, on="JobRole")
rotacion["Tasa_%"] = (rotacion["Bajas"] / rotacion["Total"] * 100).round(1)
rotacion = rotacion.sort_values("Tasa_%", ascending=False)

rotacion
```

**Interpretaciones clave**

- Sales Representative es el puesto más crítico — casi el 40% abandona la empresa. Es un rol orientado al cliente y clave para la expansión europea, lo que lo hace doblemente preocupante.
- Human Resources y Laboratory Technician también superan el 20% de rotación — niveles muy altos que merecen atención.
- Research Director y Manager son los puestos más estables, con tasas por debajo del 5% — perfiles senior con mayor vinculación a la empresa.
- Para el análisis del curso de idiomas: tiene poco sentido invertir en Sales Representative si casi el 40% se va. Habría que cruzarlo con otros filtros.

### PREGUNTA 4: ¿La satisfacción laboral influye en la salida?  JobSatisfaction, EnvironmentSatisfaction, Attrition  Evaluar impacto de la satisfacción. 

```python
# Ver satisfacción de quienes se fueron:
df[df["Attrition"] == "Yes"]["JobSatisfaction"].value_counts().reset_index()
```

```python
#dato en porcentaje:
df[df["Attrition"] == "Yes"]["JobSatisfaction"].value_counts(normalize=True) * 100
```

```python
#tenemos 29 nulos:
df["JobSatisfaction"].isnull().sum()
```

```python
# PREGUNTA 4. ¿LA SATISFACCIÓN LABORAL INFLUYE EN LA SALIDA? (combinado con lo anterior, para ver si hay relación entre puesto, satisfacción y rotación)

result = df.groupby("Attrition")["JobSatisfaction"].value_counts(normalize=True).mul(100).round(1).unstack()
result.columns = [f"Nivel {int(c)}" for c in result.columns]

result
```

```python
# PREGUNTA 4. ¿LA SATISFACCIÓN LABORAL INFLUYE EN LA SALIDA?

df.groupby("Attrition")[["JobSatisfaction", "EnvironmentSatisfaction"]].mean().round(2)
```

**Conclusiones:**
1. Los que se van tienen menor satisfacción en ambas métricas
Tanto en satisfacción laboral como en el entorno, los empleados que abandonan puntúan consistentemente por debajo de los que se quedan.
2. La diferencia no es dramática pero es consistente
~0.3 puntos de diferencia en ambas variables. No es un abismo, pero la consistencia entre las dos métricas refuerza que la satisfacción sí es un factor que influye.
3. Ningún grupo está realmente satisfecho
La escala es del 1 al 4 y los que se quedan solo puntúan 2.78 — esto indica un problema general de satisfacción en la empresa, no solo entre los que se van.

### PREGUNTA 5:¿Las horas extra están relacionadas con mayor rotación? | OverTime, Attrition | Ver si trabajar más horas aumenta la salida. 

```python
# PREGUNTA 5. ¿LAS HORAS EXTRA ESTÁN RELACIONADAS CON MAYOR ROTACIÓN?

overtime = df.groupby("OverTime")["Attrition"].value_counts().unstack(fill_value=0)
overtime["Tasa de Rotación (%)"] = (overtime["Yes"] / (overtime["Yes"] + overtime["No"]) * 100).round(1)
overtime = overtime.reset_index()
overtime.columns = ["Horas Extra", "Se quedaron", "Se fueron", "Tasa de Rotación (%)"]

overtime
```

**Interpretaciones:**
1. Las horas extra triplican la rotación
Empleados con horas extra tienen una tasa del 30.9% frente al 10.2% de los que no las hacen.
2. Es una de las variables más impactantes del dataset
Una diferencia de 20 puntos porcentuales es muy significativa — pocas variables tendrán tanto peso como esta.
3. Coste oculto para ABC Corporation
Las horas extra pueden parecer productivas a corto plazo, pero si 1 de cada 3 empleados acaba marchándose, el coste real de reemplazarlos supera con creces el beneficio.
4. Conecta con la pregunta anterior
Más horas → menos equilibrio trabajo-vida → menor satisfacción → mayor rotación. Es una cadena de causa-efecto que refuerza los resultados de la pregunta 4.

### PREGUNTA 6: ¿La distancia desde casa afecta a la retención? | DistanceFromHome, Attrition | Analizar si vivir lejos influye en irse. 

```python
# PREGUNTA 6. ¿LA DISTANCIA DESDE CASA AFECTA A LA RETENCIÓN?

df.groupby("Attrition")["DistanceFromHome"].mean().round(2)
```

**Conclusiones:**
1. Los que se van viven más lejos
Una diferencia de casi 2 km de media — los que abandonan viven a 10.68 km frente a los 8.91 de los que se quedan.
2. El impacto existe pero es moderado
No es una diferencia dramática. La distancia influye pero probablemente como factor acumulativo junto a otros como horas extra o baja satisfacción, no por sí sola.
3. La mediana es más reveladora que la media
La mediana de los que se van es 9 km frente a 7 km de los que se quedan — confirma el patrón sin que los valores extremos distorsionen.

### PREGUNTA 7. ¿El salario mensual influye en la rotación? | MonthlyIncome, Attrition | Comparar ingresos entre quienes se quedan y se van. |

```python
# PREGUNTA 7. ¿EL SALARIO MENSUAL INFLUYE EN LA ROTACIÓN?

df.groupby("Attrition")["MonthlyIncome"].describe().round(2)
```

La mediana de los que se van es 3.187$ frente a 5.205$ de los que se quedan. Casi 2.000$ de diferencia — el patrón es claro y consistente.
3. El salario bajo es un factor de riesgo claro
Junto con las horas extra y la baja satisfacción, el salario bajo forma el núcleo de los factores que empujan a los empleados a irse.
4. Implicación para ABC Corporation
Revisar la política salarial, especialmente en los puestos con mayor rotación como Sales Representative y Laboratory Technician, podría tener un impacto directo en la retención.
