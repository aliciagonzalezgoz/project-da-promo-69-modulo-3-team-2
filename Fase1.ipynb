```python
## Análisis detallado de valores nulos

#Calculamos exactamente cuántos nulos hay en cada columna y qué porcentaje representan sobre el total. Un porcentaje alto de nulos puede indicar que esa columna no es fiable.

nulos = df.isnull().sum()
pct   = (nulos / len(df) * 100).round(2)

resumen_nulos = pd.DataFrame({
    'Nulos': nulos,
    'Porcentaje (%)': pct
}).sort_values('Nulos', ascending=False)

print("Columnas CON nulos:")
print(resumen_nulos[resumen_nulos['Nulos'] > 0].to_string())
print(f"\nColumnas SIN nulos: {(nulos == 0).sum()}")
print(f"Total de valores nulos en el dataset: {nulos.sum()}")
```

**Interpretación:**
- Ninguna columna supera el 12% de nulos, así que no hay que descartar ninguna.
- `StandardHours` tiene nulos pero como es una columna redundante se eliminará igualmente.
- Los nulos se imputarán en la Fase 2: **mediana** para columnas numéricas (más robusta que la media ante valores extremos) y **moda** para columnas categóricas (el valor más frecuente).


## Variable objetivo: Attrition

`Attrition` es la columna más importante del dataset. Indica si el empleado abandonó la empresa (`Yes`) o sigue en ella (`No`). 

Para el proyecto de cursos de idiomas, **solo nos interesan los empleados con `Attrition = No`**. Ofrecer un curso caro a alguien que ya se fue o que tiene alta probabilidad de irse sería un desperdicio de recursos.

```python
print("=== DISTRIBUCIÓN DE ATTRITION ===\n")
print(df['Attrition'].value_counts())
print()
print("Porcentaje:")
print((df['Attrition'].value_counts(normalize=True) * 100).round(1).astype(str) + '%')
print()

# Universo de candidatos potenciales
candidatos = df[df['Attrition'] == 'No']
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
print(f"Total empleados en el dataset:       {len(df):>5}")
print(f"Empleados que se fueron (excluidos): {df['Attrition'].eq('Yes').sum():>5}")
print(f"Candidatos potenciales al curso:     {len(candidatos):>5}")
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
```

**Interpretación:**
- El **16,1%** de los empleados se fue de la empresa — esto es la tasa de rotación.
- El **83,9%** sigue en la empresa → **1.236 candidatos potenciales** antes de aplicar cualquier otro filtro.
- A partir de aquí, todos los análisis de perfil se harán sobre este subconjunto.
