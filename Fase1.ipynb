### PREGUNTA3. ¿Qué puestos tienen más riesgo de rotación?  JobRole, Attrition  Identificar roles con mayor salida. 
 
- Ver todos los puestos con más rotación, para saber cuales tiene más tasa de abandono:

```python
#para calcular la tasa de rotación por puesto, necesitamos el número de empleados que se fueron por cada puesto (absolutos) y el número total de empleados por cada puesto (total).
absolutos = df[df["Attrition"] == "Yes"]["JobRole"].value_counts().reset_index()
absolutos.columns = ["JobRole", "Bajas"]

# Total de empleados por puesto
total = df["JobRole"].value_counts().reset_index()
total.columns = ["JobRole", "Total"]

# Unir y calcular tasa de rotación
rotacion = absolutos.merge(total, on="JobRole")
rotacion["Tasa_%"] = (rotacion["Bajas"] / rotacion["Total"] * 100).round(1)
rotacion = rotacion.sort_values("Tasa_%", ascending=False)

rotacion
```

**Interpretaciones clave**

- Sales Representative es el puesto más crítico — casi el 40% abandona la empresa. Es un rol orientado al cliente y clave para la expansión europea, lo que lo hace doblemente preocupante.
- Human Resources y Laboratory Technician también superan el 20% de rotación — niveles muy altos que merecen atención.
- Research Director y Manager son los puestos más estables, con tasas por debajo del 5% — perfiles senior con mayor vinculación a la empresa.
- Para el análisis del curso de idiomas: tiene poco sentido invertir en Sales Representative si casi el 40% se va. Habría que cruzarlo con otros filtros.

### PREGUNTA 5:¿Las horas extra están relacionadas con mayor rotación? | OverTime, Attrition | Ver si trabajar más horas aumenta la salida. 

```python
# PREGUNTA 5. ¿LAS HORAS EXTRA ESTÁN RELACIONADAS CON MAYOR ROTACIÓN?

overtime = df.groupby("OverTime")["Attrition"].value_counts().unstack(fill_value=0)
overtime["Tasa de Rotación (%)"] = (overtime["Yes"] / (overtime["Yes"] + overtime["No"]) * 100).round(1)
overtime = overtime.reset_index()
overtime.columns = ["Horas Extra", "Se quedaron", "Se fueron", "Tasa de Rotación (%)"]

overtime
```

**Interpretaciones:**
1. Las horas extra triplican la rotación
Empleados con horas extra tienen una tasa del 30.9% frente al 10.2% de los que no las hacen.
2. Es una de las variables más impactantes del dataset
Una diferencia de 20 puntos porcentuales es muy significativa — pocas variables tendrán tanto peso como esta.
3. Coste oculto para ABC Corporation
Las horas extra pueden parecer productivas a corto plazo, pero si 1 de cada 3 empleados acaba marchándose, el coste real de reemplazarlos supera con creces el beneficio.
4. Conecta con la pregunta anterior
Más horas → menos equilibrio trabajo-vida → menor satisfacción → mayor rotación. Es una cadena de causa-efecto que refuerza los resultados de la pregunta 4.

### PREGUNTA 10: ¿Existe diferencia en la rotación según el género? | Gender, Attrition | Detectar si hay disparidad en la salida entre hombres y mujeres.

```python
# PREGUNTA 10. ¿EXISTE DIFERENCIA EN LA ROTACIÓN SEGÚN EL GÉNERO?

genero = df.groupby("Gender")["Attrition"].value_counts().unstack(fill_value=0)
genero["Tasa_%"] = (genero["Yes"] / (genero["Yes"] + genero["No"]) * 100).round(1)
genero.columns = ["Se quedaron", "Se fueron", "Tasa_%"]

genero
```

**Conclusiones:**
1. Los hombres tienen una tasa de rotación ligeramente mayor
17.1% frente al 14.8% de las mujeres — una diferencia de 2.3 puntos porcentuales.
2. La diferencia existe pero no es alarmante
No hay una disparidad grave por género. Ambos grupos tienen tasas relativamente similares, lo que sugiere que el género por sí solo no es un factor determinante.
3. Convendría cruzarlo con otras variables
Para entender mejor si hay algún patrón oculto, sería interesante cruzar género WorkLifeBalance y Attrition

```python
# PREGUNTA 10 CRUZADA CON WORKLIFEBALANCE

resultado = df.groupby(["Gender", "WorkLifeBalance"])["Attrition"].value_counts(normalize=True).mul(100).round(1).unstack(fill_value=0)
resultado.columns = ["Se quedaron (%)", "Se fueron (%)"]

resultado
```

### PREGUNTA 11: ¿Qué variables necesitan limpieza antes del análisis? | EmployeeCount, Over18, StandardHours + nulos/duplicados | Identificar columnas constantes y problemas de calidad. |

```python
# Columnas constantes
constantes = [col for col in df.columns if df[col].nunique() == 1]
pd.DataFrame({
    "Columna": constantes,
    "Valor único": [df[col].unique()[0] for col in constantes],
    "Acción": ["Eliminar en Fase 2"] * len(constantes)
})
```

```python
# Nulos por columna ordenados
df.isnull().sum()[df.isnull().sum() > 0].sort_values(ascending=False)
```

```python
# Duplicados
df.duplicated().sum()
```

```python
def eda_basico(df):
    print("🔍 Primeras filas del DataFrame:")
    display(df.head())

    print("📐 Dimensiones:")
    print(f"{df.shape[0]} filas y {df.shape[1]} columnas\n")

    print("🧠 Información general:")
    display(df.info())

    print("📊 Tipos de datos por columna:")
    display(df.dtypes.rename("Tipo").to_frame())

    print("📉 Descripción de columnas numéricas:")
    display(df.describe())

    print("🔤 Descripción de columnas categóricas:")
    display(df.describe(include=["O"]))

    print("🚫 Valores nulos por columna:")
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0].sort_values(ascending=False)
    if len(nulos) > 0:
        display(pd.DataFrame({
            "Nulos": nulos,
            "Porcentaje (%)": (nulos / len(df) * 100).round(2)
        }))
    else:
        print("No hay valores nulos.\n")

    print("🗑️ Columnas constantes:")
    constantes = [col for col in df.columns if df[col].nunique() == 1]
    if constantes:
        display(pd.DataFrame({
            "Columna": constantes,
            "Valor único": [df[col].unique()[0] for col in constantes],
            "Acción": ["Eliminar en Fase 2"] * len(constantes)
        }))
    else:
        print("No hay columnas constantes.\n")

    print("📎 Filas duplicadas:")
    dup_count = df.duplicated().sum()
    print(f"Duplicadas: {dup_count}")
    if dup_count > 0:
        print("Ejemplo de duplicados:")
        display(df[df.duplicated()].head())
    else:
        print("No hay filas duplicadas.\n")
```

```python
eda_basico(df)
```
