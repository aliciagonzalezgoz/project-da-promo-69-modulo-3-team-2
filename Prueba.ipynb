# 📊 Preguntas clave para el análisis de rotación

| Nº | Pregunta | Variables involucradas | Objetivo del análisis |
|----|----------|------------------------|------------------------|
| 1 | ¿Qué porcentaje de empleados abandona la empresa? | Attrition | Medir la rotación general. |
| 2 | ¿La rotación cambia según departamento? | Department, Attrition | Comparar rotación por áreas. |
| 3 | ¿Qué puestos tienen más riesgo de rotación? | JobRole, Attrition | Identificar roles con mayor salida. |
| 4 | ¿La satisfacción laboral influye en la salida? | JobSatisfaction, EnvironmentSatisfaction, Attrition | Evaluar impacto de la satisfacción. |
| 5 | ¿Las horas extra están relacionadas con mayor rotación? | OverTime, Attrition | Ver si trabajar más horas aumenta la salida. |
| 6 | ¿La distancia desde casa afecta a la retención? | DistanceFromHome, Attrition | Analizar si vivir lejos influye en irse. |
| 7 | ¿El salario mensual influye en la rotación? | MonthlyIncome, Attrition | Comparar ingresos entre quienes se quedan y se van. |
| 8 | ¿La antigüedad reduce o aumenta la rotación? | YearsAtCompany, TotalWorkingYears, YearsInCurrentRole | Ver cómo influye la experiencia y permanencia. |
| 9 | ¿La falta de promoción está relacionada con la salida? | YearsSinceLastPromotion, Attrition | Evaluar si no promocionar aumenta la rotación. |
